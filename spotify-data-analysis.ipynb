```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_tracks.info()
```

```python
sorted_data = data_tracks.sort_values('popularity', ascending=True).head(10)
sorted_data
```

```python
data_tracks.describe().transpose()
```

```python
popular_songs= data_tracks.query('popularity >90', inplace= False).sort_values('popularity',ascending= False)
popular_songs[:10]
```

```python
data_tracks.set_index("release_date", inplace= True)
data_tracks.index=pd.to_datetime(data_tracks.index)
data_tracks.head()
```

```python
data_tracks[["artists"]].iloc[18]
```

```python
data_tracks["duration"]= data_tracks["duration_ms"].apply(lambda x :round(x/1000))
data_tracks.drop("duration_ms",inplace=True,axis =1)
```

```python
data_tracks.duration.head()
```

```python
corr_df = data_tracks.drop(["key",'mode','explicit'],axis=1).corr(method="pearson")
plt.figure(figsize= (14,6))
heatmap= sns.heatmap(corr_df,annot= True,fmt= ".1g", vmin=-1,vmax=1, center=0,cmap="inferno",linewidths=1,linecolor='black')
heatmap.set_title("correaltion Heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
sample_data=data_tracks.sample(int(0.004*len(data_tracks)))
print(len(sample_data))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y='loudness',x='energy',color='c').set(title="loudness vs energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y='popularity',x='liveness',color='c').set(title="popularity vs liveness")
```

```python
data_tracks["dates"]=data_tracks.index.get_level_values('release_date')
data_tracks.dates=pd.to_datetime(data_tracks.dates)
years= data_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="number of songs per year")
```

```python
total_dr = data_tracks.duration 
fig_dims=(18,7)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x= years, y = total_dr, ax= ax,errwidth=False).set(title= "year vs duration ")
plt.xticks(rotation=90)
```
